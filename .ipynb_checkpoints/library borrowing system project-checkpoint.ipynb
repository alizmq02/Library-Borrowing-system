<h2>Function-1</h2>
<h3>getAllRecords()</h3>
<p>This function will display all members' records</p>

```python
def getAllRecords():
    try: #The program use try and except in order if the file does not exsist it prinst an appropriate error.
        inputFile = open("LMS.txt","r") #the program open the file to read it.
        line = inputFile.readline() # here the program read the first line then the second because the first line have the titles which the program will display it using print function.
        line = inputFile.readline()
        print("  %-15s%-21s%-20s%-10s%s" % ("Member ID", "Member Name", "Borrowed Books", "Visits", "Books Per Visit"))#Thr program here print the titles bar
        while line != "" : #the program starts while to read and display the lines' information each one alone until the end of the lines and the program will stop the while loop when line value become empty string which mean that there are no lines to read.    
            member = line.split() #make the line information as a list so the program can deal with it easily.
            memberId = member[0] # the program here separate each informaition in the list in a variable
            memberName = member[1]+" "+member[2]
            borrowedBooks = member[3]
            visits = member[4]
            booksPerVisit = member[5]
            print("  %-15s%-27s%-16s%-13s%s" % (memberId, memberName, borrowedBooks, visits, booksPerVisit)) #here the program print the member information
            line = inputFile.readline() #the program update line value to the next line so the loop work until it finishs the lines.
        inputFile.close()
    except IOError:
        print("Error: file not found")
    return 0
```

<h2>Function-2</h2>

```python
def getMemberRecord(memberID):
    try: # the program use try and excipt in order if the file does not exsist it prints an appropriate error.
        inputFile = open("LMS.txt","r") #open the file to read it
        line = inputFile.readline() #the programe read the first line then updatee the variable line to the second because the first line is just the titles bar. 
        line = inputFile.readline()
        check = 0 # the program define check , checkCondition , checklen and originLen because it will use them in if conditions below to make sure that the values entered are correct.
        checkCondition = 0
        checkLen = len(memberID)
        originLen = 4
        while line != "" : # the program starts a while loop in order to read all the lines thaqt have an information until the end of the file.
            member = line.split() #put lines information in a list in order to deal with it easily  
            if checkLen == originLen and (memberID in member[0]): #here it put these if condition to make sure that all the variables that the user enters are correct and exsit in the file.
                memberId = member[0] #here the program put the list information in a variables in order to deal with them easily
                memberName = member[1]+" "+member[2]
                borrowedBooks = member[3]
                visits = member[4]
                booksPerVisit = member[5]
                if check == checkCondition: #the program put this condition in order to print titles bar once then the check value will update so that it does not write it again.
                    print("  %-15s%-21s%-20s%-10s%s" % ("Member ID", "Member Name", "Borrowed Books", "Visits", "Books Per Visit"))
                    check = check + 1 #here it update the value of check
                print("  %-15s%-27s%-16s%-13s%s" % (memberId, memberName, borrowedBooks, visits, booksPerVisit,)) # here it prints the member information.
            line = inputFile.readline() #it updates the value of line in order to run the while until it reavhes the end of the lines that have information.
        if check == checkCondition : # here if check and check condition remain equal that means that the values that the user enter does not enter the if condition which means that the id that the user enter is not correct so it will print error massage. 
            print("Error: Invalid Member ID")
        inputFile.close()
    except IOError:
        print("Error: file not found")
    return 0
```

<h2>Function-3</h2>
<h3>updateMemberRecord(memberID)</h3>
<p>This function will update a member information.</p>

```python
def updateMemberRecord(memberID, bookCount):
    bookCount1 = int(bookCount)
    condition = 0
    condition2 = 0
    inputFile = open("LMS.txt","r") # open the file that the program will edit the information in it
    line = inputFile.readline() #the program read the first kine then the second because the first line is just titles bar  
    line = inputFile.readline()
    #the program define this variable in order to append some information in it.
    membersList = []
    while line != "": # the programe set a while loop to save all lines information
        lineList = line.split() # the program make the line as a list in order to deal with it easily.
        membersList.append(lineList) #appends all lines' information.
        line = inputFile.readline() # read the next line in order to run while loop until the end of the lines.
    inputFile.close() #close the file after finish using it.
    outputFile = open("LMS.txt", "w") # after the programe make sure that all variables values are correct, it starts writing the iformations.
    for i in membersList: # membersList has lists inside a list and each list have information of a member,so the program use for i in order to take each list from the main list and make the informations inside it as variables ,snd this will allow the program write and deal with the informations easily.
        memberId1 = i[0]
        memberName = i[1] + " " + i[2]
        borrowedBooks = i[3]
        visits = int(i[4]) # the program make it integer in order to make it updatable. 
        booksPerVisit = float(i[5]) # the program make it float in order to make it updatable.
        if memberId1 == memberID:  # here the program set a condition that says if the id in these list same as the id that enterd, update the visits and the book per visit and the borrowed variables                   
            booksPerVisit = (bookCount1 + (booksPerVisit * visits)) / (visits + 1)
            visits = visits + 1
            borrowedBooks = bookCount
        if condition2 == condition: # the program use this condition so it writes the titles bar 
            outputFile.write("  %-15s%-21s%-20s%-10s%s" % ("Member ID", "Member Name", "Borrowed Books", "Visits", "Books Per Visit"+"\n"))
            condition2 = condition2 + 1 #the program update condition2 value so when the loop run again it does not write titles bar again 
        outputFile.write("  %-15s%-27s%-16s%-13s%.2f" % (memberId1, memberName, borrowedBooks, visits, booksPerVisit) + "\n")  # here the program write the information in the file                   
    outputFile.close()
    print()
    print()
    print("Member's information has been updated...")
    return 0
#it also use this function to make sure that the function above take correct variables.
def checkIfBookCountValid(bookCount):
    bookCount1 = int(bookCount)
    condition2 = 4
    condition1 = 0
    if len(bookCount) > condition2 : #the program uses this conditions in order to make sure that the value entered are correct
        print("Error: Members can take up to five books in one visit")
        return False
    elif bookCount1 < condition1 :
        print("Error: Invalid No. of book to borrow")
        return False
    else :
        return True 
```
